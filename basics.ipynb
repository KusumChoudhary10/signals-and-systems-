# Basics of matplotlib library in python

#### importing necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

1.basic graph :

```python
x = [0,1,2,3]
y = [0,2,4,6]

#plt.plot(x,y, label='2x', color='skyblue', linewidth='2',linestyle='--',marker='.',markersize='15')
plt.plot(x,y, 'bo-.',label='2x')

#2nd line
x2= np.arange(0,2,0.3)
plt.plot(x2[:6], x2[:6]**3,'r')
plt.plot(x2[5:],x2[5:]**3,'r.:')

plt.title("Basic Graph", fontdict={'fontname' : 'Sans', 'fontsize':'20'})
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.xticks([0,1,2,3,4,5,])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

#### **functions:-**\
**1.xlabel/ylabel**: used to name the axes.\
**2.plt.title()**: for adding a title to the graph.\
**3.plt.show()**: shows the graph without the scaling.\
**4.xticks/yticks**: can be used to specify the values on the axes.
